3º Exercício: String<br>
3.1 - Aplique três métodos distintos na *string* abaixo:

cancao = 'Roda mundo, roda gigante, roda moinho, roda pião.'

3.2 - Extraia da string abaixo o valor da taxa **selic** na variável `selic` e o valor do **ano** na variavel `ano`. Imprima os valores na tela.

noticia = 'Selic vai a 2,75% e supera expectativas; é a primeira alta em 6 anos.'


```python
# Exercicio 3
cancao = 'Roda mundo, roda gigante, roda moinho, roda pião.'

#3.1
# Metodo 1 upper() - Converte uma cadeia de caracteres em maiúsculas
x = cancao.upper()
print('Saida UPPER: '+x)

# Metodo 2 count() - Retorna o número de vezes que um valor especificado ocorre em uma cadeia de caracteres
y = cancao.count("roda")
print(y)

# Metodo 3 casefold() - Converte a cadeia de caracteres em minúsculas
z = cancao.casefold()
print('Saida Casefold: '+z)

# Metodo 4 find() - Pesquisa a cadeia de caracteres por um valor especificado e retorna a posição de onde ela foi encontrada
x = cancao.find("mundo")
print(x)

# Metodo 5 isnumeric() - Retorna True se todos os caracteres na cadeia de caracteres forem numéricos
x = cancao.isnumeric()
print(x)

# 3.2
#Extraia da string abaixo o valor da taxa **selic** na variável `selic`
# e o valor do **ano** na variavel `ano`. Imprima os valores na tela.
noticia = 'Selic vai a 2,75% e supera expectativas; é a primeira alta em 6 anos.'

#Extraindo valor da taxa selic na variavél noticia e armazenando na variavél selic - Posição 11 : 16
selic = noticia[11:17]
print('Taxa Selic:'+selic)

#Extraindo valor do ano na variavél noticia e armazenando na variavél ano - Posição 62
ano = noticia[62]
print('Ano: '+ano)

#Fim exercicio 3


```
